```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("country_wise_latest.csv")
df.head()
```

# 1. BAR GRAPH

```python
plt.bar(df["Country/Region"][::5], df.Confirmed[::5])

plt.xlabel("Country/Region")
plt.ylabel('Confirmed')
plt.xticks(rotation=90)
plt.plot()
```

```python
# It helps us identify the countires with the highest number of confirmed Covid cases
# Here we can see that Chile, Saudhi Arabia and Italy have the most confirmed cases
```

# 2. LINE GRAPH

```python
plt.plot(df["Country/Region"][::5], df["1 week change"][::5])

plt.xlabel("Country/Region")
plt.ylabel('1 week change')
plt.xticks(rotation=90)
plt.plot()
```

```python
# It helps us identify the countires with the highest change in rate of cases
# Saudi Arabia has the highest rate of change in number of Covid cases and Chile has the second highest change rate
```

# 3. Pie Char

```python
df1 = df.groupby("WHO Region").count()
plt.pie(df1["Confirmed"], labels=df1.index, autopct='%1.1f%%')
plt.show()
```

```python
# Of total cases Europe contributes ot 29.9% of cases and Africa contributes to 25.7% of the cases
```

# 4. SCATTER

```python
df2 = df.sort_values("Deaths")

plt.scatter(df.Active[::2], df.Deaths[::2])
plt.xlabel("Active")
plt.ylabel("Deaths")
plt.show()
```

```python
# Scatter plot that shows the relationship between active cases and deaths. 
# This can help identify if higher active cases correlate with higher death numbers
```

# 5. HISTOGRAM

```python
sns.histplot(df["New cases"][::2])
plt.show()
```

```python
# Shows Distribution of new cases. Helps see the frequency of new case counts
```

# 6. Heat Map

```python
data = np.random.rand(10,10)
plt.imshow(data, cmap='coolwarm', interpolation='nearest')
plt.colorbar(label='')
plt.show()
```

# 7. Blox Plot

```python
q1=np.percentile(df["New deaths"],25)
q3=np.percentile(df["New deaths"],75)
iqr=q3-q1

uw=q3+1.5*iqr
lw=q1-1.5*iqr

plt.boxplot(df["New deaths"][(df["New deaths"]>lw)&(df["New deaths"]<uw)])
plt.ylabel("New Deaths")
plt.show()
```

```python
# It Displays new deaths across regions. Also helps us identify ouliners.
```

# 8. Area Graph

```python
plt.plot(df["Country/Region"][::5], df["1 week change"][::5])


plt.xlabel("Country/Region")
plt.ylabel('Confirmed')
plt.xticks(rotation=90)
plt.plot()
plt.fill_between(df["Country/Region"][::5], df["1 week change"][::5], color="teal", alpha=0.4)
```

```python
# It helps us identify the countries with the highest number of confirmed cases
```

# 9. Stem

```python
plt.stem(df["Country/Region"][::5], df["1 week change"][::5])
plt.xlabel("Country/Region")
plt.ylabel('Confirmed')
plt.xticks(rotation=90)
plt.plot()
```

# 10. Stack Plot

```python
Active = df["Active"][::5]
Deaths = df["Deaths"][::5]
Recovered = df["Recovered"][::5]
#This is done to apply label, so that legend can show appropriate name
plt.plot([], [], color ='red', 
         label ='Active')
plt.plot([], [], color ='green',
         label ='Deaths')
plt.plot([], [], color ='blue',
         label ='Recovered')

plt.stackplot(df["Country/Region"][::5], Active ,Deaths, Recovered, colors=['r','g','b'])
plt.xlabel("Country/Region")
plt.ylabel('Confirmed, Deaths, Recovered')
plt.xticks(rotation=90)
plt.plot()
plt.legend()
```

```python
# It helps us correlate the between the number of Confirmed cases, deaths, and recovered cases on the same graph
```
